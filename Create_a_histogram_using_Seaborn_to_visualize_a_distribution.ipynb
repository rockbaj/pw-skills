## 5. Create a Histogram using Seaborn

We‚Äôll visualize the **distribution of insurance charges**.

```python
# Set a basic Seaborn style
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))

sns.histplot(
    data=df,
    x="charges",
    bins=30,       # number of bars
    kde=True       # add KDE curve (smooth density line)
)

plt.title("Distribution of Insurance Charges")
plt.xlabel("Charges")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

## 6. Interpretation (for your write-up)

* The **x-axis** shows the insurance `charges`.
* The **y-axis** shows how many people fall into each charge range.
* The **histogram bars** show the frequency distribution.
* The **KDE curve** (smooth line) helps us see the overall shape of the distribution
  (often right-skewed for cost/charges data).

````python
import kagglehub
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
````

````python
# Download latest version of the Insurance dataset
path = kagglehub.dataset_download("mirichoi0218/insurance")

print("Dataset path:", path)
print("Files in folder:", os.listdir(path))
````

````python
csv_path = os.path.join(path, "insurance.csv")

# Load dataset
df = pd.read_csv(csv_path)

# Preview data (optional)
df.head()
````

````python
# Set a basic Seaborn style
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))

sns.histplot(
    data=df,
    x="charges",
    bins=30,       # number of bars
    kde=True       # add KDE curve (smooth density line)
)

plt.title("Distribution of Insurance Charges")
plt.xlabel("Charges")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
````
