Q 3:Explain the difference between a Z-test and a T-test, including when to use
each.
`

-- The Z-test and T-test are both statistical hypothesis tests used to compare means. The primary difference lies in the information you have about the population standard deviation and the sample size. Here's a breakdown:

Z-test

Purpose: Used to determine if there is a significant difference between a sample mean and a population mean, or between two sample means.
Assumptions:
Known Population Standard Deviation (σ): This is the most crucial assumption. You must know the true standard deviation of the population from which the sample is drawn.
Large Sample Size (n > 30): Even if the population standard deviation is unknown, if the sample size is sufficiently large (typically n > 30), the sample standard deviation can be a good estimate of the population standard deviation, and the Z-test can still be approximately used (due to the Central Limit Theorem).
Normally Distributed Data: The population from which the sample is drawn should be normally distributed, or the sample size should be large enough for the Central Limit Theorem to apply (making the sampling distribution of the mean approximately normal).
Formula (for one-sample Z-test): $$ Z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}} $$ Where:
x̄ is the sample mean
μ is the population mean
σ is the population standard deviation
n is the sample size
When to Use:
When the population standard deviation is known.
When the sample size is large (n > 30), even if the population standard deviation is unknown (as the sample standard deviation can be used as a reasonable estimate for σ).
T-test

Purpose: Used to determine if there is a significant difference between a sample mean and a population mean, or between two sample means, particularly when the population standard deviation is unknown.
Assumptions:
Unknown Population Standard Deviation (σ): This is the key distinguishing factor. 

        The T-test uses the sample standard deviation (s) as an estimate for the population standard deviation.
Small Sample Size (n < 30): The T-test is especially appropriate for smaller sample sizes where the sample standard deviation might not be a very precise estimate of the population standard deviation.
Normally Distributed Data: The population from which the sample is drawn should be approximately normally distributed.
Formula (for one-sample T-test): $$ t = \frac{\bar{x} - \mu}{s / \sqrt{n}} $$ Where:
x̄ is the sample mean
μ is the hypothesized population mean
s is the sample standard deviation
n is the sample size
When to Use:
When the population standard deviation is unknown.
When the sample size is small (n < 30).
When comparing two means (independent samples or paired samples) and the population standard deviations are unknown.
Key Differences Summarized:

Feature	Z-test	T-test
Population Std Dev (σ)	Known	Unknown (estimated by sample std dev, s)
Sample Size (n)	Generally large (n > 30)	Can be small (n < 30) or large
Distribution Used	Standard Normal (Z) distribution	Student's t-distribution (with degrees of freedom)
When to Use Primarily	When σ is known, or large n	When σ is unknown
In practice:

In many real-world scenarios, the population standard deviation is unknown. Therefore, the T-test is much more commonly used than the Z-test. The Z-test is more theoretical or used in situations where there's extensive historical data or a very large sample size that makes the sample standard deviation a highly reliable estimate of the population standard deviation.
