`
Write a Python program that reads from one file and writes its content to
another file
`

```python
input_file_name = 'input.txt'
output_file_name = 'output.txt'

# Create a dummy input file for demonstration
# In a real scenario, this file would already exist
try:
    with open(input_file_name, 'w') as f:
        f.write('This is the content of the input file.\n')
        f.write('Line 2: Hello, world!\n')
        f.write('Line 3: Python file operations are useful.\n')
    print(f"Dummy input file '{input_file_name}' created successfully.")
except IOError as e:
    print(f"Error creating input file: {e}")

# Read from the input file and write to the output file
try:
    with open(input_file_name, 'r') as infile:
        content = infile.read()

    with open(output_file_name, 'w') as outfile:
        outfile.write(content)

    print(f"Content successfully copied from '{input_file_name}' to '{output_file_name}'.")

except FileNotFoundError:
    print(f"Error: The file '{input_file_name}' was not found.")
except IOError as e:
    print(f"An I/O error occurred: {e}")

# Optional: Verify the content of the output file
try:
    with open(output_file_name, 'r') as f:
        print(f"\nContent of '{output_file_name}':")
        print(f.read())
except FileNotFoundError:
    print(f"Error: The output file '{output_file_name}' was not found after writing.")
except IOError as e:
    print(f"An I/O error occurred when reading output file: {e}")
```
